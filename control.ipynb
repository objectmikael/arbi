```python
import os
import pandas as pd
from datetime import datetime
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
# Load environment variables from .env file
load_dotenv()

# Access the variables 
psql_username = os.getenv('PSQL_USERNAME')
psql_password = os.getenv('PSQL_PASSWORD')
psql_host = os.getenv('PSQL_HOST')
psql_port = os.getenv('PSQL_PORT')
db_name = os.getenv('CONTROL_DB_NAME')

# Define the database url
db_url = f"postgresql://{psql_username}:{psql_password}@{psql_host}:{psql_port}/{db_name}" 

# Create the engine object
engine = create_engine(db_url)

# Write ethe SQL query
query = 'SELECT * FROM trades'

# Read the SQL query into a DataFrame
control_df = pd.read_sql(query, engine)

# Show the DataFrame's head
control_df
```

```python
control_df['current_datetime'] = pd.to_datetime(control_df['current_datetime'])
control_df['current_datetime'] = pd.to_datetime(control_df['current_datetime'].dt.date)

control_df.set_index('current_datetime')
display(control_df.groupby(['currency', 'buy_exchange'])[['total_purchase_amount','profit', 'spread_percentage', 'wallet_balance']].mean())
display(control_df.groupby(['currency'])[['total_purchase_amount','profit', 'spread_percentage', 'wallet_balance']].mean())
```
